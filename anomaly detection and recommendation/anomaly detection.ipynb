```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.io as sio
import math
import seaborn as sns
sns.set(context="notebook", style="white", palette=sns.color_palette("RdBu"))
```

```python
mat = sio.loadmat('data/ex8data1.mat') 
data = pd.DataFrame(mat['X'], columns=['X1', 'X2'])
data.head()
```

```python
sns.regplot(data['X1'], data['X2'], x='Latency', y='Throughput', fit_reg=False)
plt.show()
```

```python
def guassianDistribution(X):

    X = np.matrix(X)
    m = len(X)
    n = X.shape[1]
    p = np.zeros((m, n))
    
    for i in range(m):
        average = np.mean(X[i, :])
        variance = np.var(X[i, :])
        p[i, :] = (1 / np.sqrt(2 * math.pi * variance)) * np.exp(-np.power((X[i, :] - average), 2) / (2 * variance)) 

    return p
```

```python
p_test = guassianDistribution(data)
p_test = pd.DataFrame(p_test, columns = ['p1', 'p2'])
sns.displot(x = data['X1'], y = p_test['p1'])
plt.show()
```
