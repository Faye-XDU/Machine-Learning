```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.io as sio
```

# Data 1

```python
mat = sio.loadmat('./data/ex7data1.mat')
data1 = pd.DataFrame(mat.get('X'), columns=['X1', 'X2'])
data1.head()
```

```python
plt.scatter(x=data1['X1'], y=data1['X2'], marker='+', c='r')
plt.show()
```
