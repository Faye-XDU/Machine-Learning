```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from skimage import io
```

```python
pic = io.imread('data/pic.jpg') / 255
io.imshow(pic)
plt.show()
```

```python
pic.shape
```

```python
data = np.reshape(pic, (pic.shape[0] * pic.shape[1], pic.shape[2]))
data.shape
```

```python
model = KMeans(n_clusters=5, max_iter=5)
model.fit(data)
```

```python
centroids = model.cluster_centers_
print(centroids.shape)
C = model.predict(data)
print(C.shape)
```

```python
centroids[C].shape
```

```python
compressed = np.reshape(centroids[C], (pic.shape[0], pic.shape[1], pic.shape[2]))
compressed.shape
```

```python
plt.imshow(compressed)
plt.show()
```
